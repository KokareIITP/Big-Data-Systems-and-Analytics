```python
# Imports
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision.models as models
import pickle
import sys
import time
```

```python
# Download the dataset - mention the specs (input size, number of images, size of each image, size of dataset)

# Get a minibatch of dataset for binary classification
def get_binary_minibatch(dataset, size, class1 = 2, class2 = 4):
    class1_count = size/2
    class2_count = size/2
    mini_dataset = []
    for data in dataset:
        
        if (class1_count <= 0 and class2_count <= 0):
            break
        
        image, label = data
        if(label == class1):
            if(class1_count <= 0 ):
                continue
            mini_dataset.append(data)
            class1_count = class1_count - 1
            

        elif(label == class2):
            if(class2_count <= 0):
                continue
            mini_dataset.append(data)
            class2_count = class2_count - 1
                
    return mini_dataset

# Get dataloaders
def get_dataloaders(trainset, testset):
    trainloader = torch.utils.data.DataLoader(trainset, batch_size=4,
                                              shuffle=True, num_workers=2)
    testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                             shuffle=False, num_workers=2)
    return trainloader, testloader

# Get outliers
def get_outliers(dataset, size, outlier_class = 0):
    outlier_count = size
    outliers = []
    for data in dataset:
        image, label = data
        if(outlier_count <= 0):
            break
        if(label == outlier_class and outlier_count > 0):
            outliers.append(data)
            outlier_count = outlier_count - 1
            
    return outliers

# Get classes
def get_classes():
    classes = ('plane', 'car', 'bird', 'cat','deer', 'dog', 'frog', 'horse', 'ship', 'truck')
    return classes
```

```python
# Shows unnormalized images
def im_show(img):
    img = img / 2 + 0.5
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# Get sample images from each class
def show_sample(dataloader, obj_class):    
    # get some random training images
    classes = get_classes()
    class_samples = []
    
    with torch.no_grad():
        for data in dataloader:
            images, labels = data
            for i in range(4):
                if(labels[i] == obj_class):
                    class_samples.append(images[i])
                    
    print('%5s ' % classes[obj_class])
    im_show(torchvision.utils.make_grid(class_samples[:5]))
```

```python
# Get the default configuration of NN
def get_NN(index):
    if(index == 0):
        return models.resnet18()
    elif(index == 1):
        return models.densenet161()
    elif(index == 2):
        return models.wide_resnet50_2()
```

```python
# Errors (Mention the default error threshold)
def get_criterion():
    return nn.CrossEntropyLoss()

def get_optimizer(net):
     return optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
# Calculate Training Accuracy and Testing Accuracy
def get_accuracy(net, testloader):
    correct = 0
    total = 0
    with torch.no_grad():
        for data in testloader:
            images, labels = data
            outputs = net(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return 100 * correct/ total

# Get testing time
def get_testing_time(net, testloader):
    start_time = time.time()
    with torch.no_grad():
        for data in testloader:
            images, labels = data
            outputs = net(images)
    
    end_time = time.time()
    return end_time - start_time
    
```

```python
# Generate dataset and dataloaders
transform = transforms.Compose(
        [transforms.ToTensor(),
         transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.CIFAR10(root='./data/cifar', train=True,
                                            download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data/cifar', train=False,
                                           download=True, transform=transform)

mini_trainset = get_binary_minibatch(trainset, 400, 2, 4)
mini_testset = get_binary_minibatch(testset, 100, 2, 4)

trainloader, testloader = get_dataloaders(mini_trainset, mini_testset)
```

```python
# Get outlier images
outliers = get_outliers(testset, 12 , 0)
outliers_loader = torch.utils.data.DataLoader(outliers, batch_size=4,
                                              shuffle=True, num_workers=2)
```

```python
# Show sample images - class1, class2, outliers

show_sample(testloader, 2)
show_sample(testloader, 4)
show_sample(outliers_loader, 0)
```

```python
def plot(y,x, location, name, ylabel, xlabel):
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_title(name)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    fig.savefig(location)
    plt.show()
    
def plot_doublebar(y,x, location, name, ylabel, xlabel):
    class1 = []
    class2 = []

    for epoch in y:
        if(not isinstance(epoch, list)):
            continue
        count1 = 0
        count2 = 0
        for entry in epoch:
            if(entry == torch.tensor(2)):
                count1 = count1 + 1
            elif(entry == torch.tensor(4)):
                count2 = count2 + 1
                
        class1.append(count1)
        class2.append(count2)
        
    fig, ax = plt.subplots()
    x_t = np.arange(len(x))
    ax.bar(x_t-0.1, class1, width=0.2, color='b', align='center', label='Bird')
    ax.bar(x_t+0.1, class2, width=0.2, color='r', align='center', label='Deer')
    ax.set_title(name)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xticks(x)
    fig.savefig(location)
    plt.show
    

def make_plots(trn_acc, tst_acc, out_acc, epochs, prefix):
    root = './results/' + prefix + '/'
    epochs = [(i + 1) for i in range(epochs)]
    
    # Training accuracy vs epochs
    plot(trn_acc, epochs, root + 'trn_acc', 'Training accuracy vs Number of Epochs', 'Training accuracy' , 'Number of Epochs')

    # Testing accuracy vs epochs
    plot(tst_acc, epochs, root + 'tst_acc', 'Testing accuracy vs Number of Epochs', 'Testing accuracy', 'Number of Epochs')

    # Outliers accuracy vs epochs
    plot_doublebar(out_acc, epochs, root + 'out_acc', 'Outliers Prediction vs Number of Epochs', 'Outliers Prediction', 'Number of Epochs')
```

```python
# Basic Model
net = get_NN(0)
    
# Results for this NN
trn_acc = []
tst_acc = []
out_acc = []


# Train NN   
print('Started Training')
training_start_time = time.time()
for epoch in range(10):  # loop over the dataset multiple times
    print('Starting Epoch %d: ' % (epoch + 1))
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer = get_optimizer(net)
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        criterion = get_criterion()
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 25 == 24:
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss /25))
            running_loss = 0.0

    # Training results
    acc1 = get_accuracy(net, trainloader)
    trn_acc.append(acc1)

    # Testing Results
    acc2 = get_accuracy(net,testloader)
    tst_acc.append(acc2)

    # Outlier Results
    pred = []
    for data in outliers_loader:
        images, labels = data
        outputs = net(images)
        _, predicted = torch.max(outputs.data, 1)
        for output in predicted:
            pred.append(output)

    out_acc.append(pred)
    
    epoch_time = time.time() - training_start_time
    
    # Get the size of the model (in MBs)
    model = pickle.dumps(net)
    byte_size = sys.getsizeof(model)
    model_size = byte_size/1000000
    
    print('End of Epoch %d: Training Accuracy: %.3f  Testing Accuracy: %.3f Epoch End Time: %.3f Model Size: %.3f' % (epoch + 1, acc1, acc2, epoch_time, model_size))

training_end_time = time.time()
print('Training Time: %.3f ' % (training_end_time - training_start_time))
print('Testing Time: %.3f'% get_testing_time(net, testloader))
print('Outlier Testing Time: %.3f'% get_testing_time(net, outliers_loader))
print('Finished Training')
```

```python
# Make Plots
make_plots(trn_acc, tst_acc, out_acc, 10, '3a')
```
